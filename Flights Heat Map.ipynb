```python
Seaborn comes with 17 built-in datasets.
Here we are gonna analyze the 'flights' dataset.
We are gonna use this data to make a hitogram.
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_flights = sns.load_dataset('flights')

df_flights
# as you can see, this dataframe don't have te best organization. We can use the .pivot() function to get a better df.
```

```python
df_flight_reshaped = df_flights.pivot('month', 'year', 'passengers')
df_flight_reshaped.head()
```

```python
fig, ax = plt.subplots(figsize=(15,10))



sns.heatmap(data = df_flight_reshaped,
                           annot=True,
                           fmt= 'd',
                          linewidths=0.01,
                          cbar = True,
                          square = False,
                          cmap='RdBu_r',
                          )

plt.title('Flights---HeatMap', fontsize=22, fontweight='bold')
plt.xlabel('Year', fontsize=16, fontweight='bold')
plt.ylabel('Month', fontsize=16, fontweight='bold', x=2)


plt.savefig('Flights.png', bbox_inches='tight', facecolor='white', transparent=False)

plt.show()
```
